## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Droppong Columns
df=df.drop(["ID","Name","Customer_ID","SSN"],axis=1)
````

````python
# Handling Fake Missing Values
df["Type_of_Loan"]=df["Type_of_Loan"].replace(["No Data","Not Specified"],np.nan)
df["Payment_of_Min_Amount"]=df["Payment_of_Min_Amount"].replace("NM",np.nan)
````

````python
# Creating a New Column
df["Loan Count"]=df["Type_of_Loan"].str.count(",")+1
````

````python
# Dropping Type_of_Loan Column
df=df.drop("Type_of_Loan",axis=1)
````

````python
# Handling NaN values in Loan Count column
df["Loan Count"]=df["Loan Count"].fillna(0)
````

````python
# Renaming Loan Count column as Loan_Count
df.rename(columns={"Loan Count":"Loan_Count"},inplace=True)
````

````python
df.isnull().sum()
````

### What all manipulations have you done and insights you found?

Manipulation Done :-       
*   Dropped irrelevant columns: ID, Customer_ID, Name, SSN.
*   Identified and converted fake missing values ("NM", "No Data", "Not Specified") into NaN.
*   Detected that Type_of_Loan had multiple values in a single row.
*  Created a new feature Loan_Count by counting number of loans.
*  Dropped Type_of_Loan after feature extraction.
*  Filled missing values in Loan_Count with 0.
*  Kept missing values in Payment_of_Min_Amount for analysis.
*  Ensured dataset is clean, consistent, and ready for EDA.

Insights so far :-                                                
*  Dataset had no duplicates and no real nulls, but contained hidden missing values.
* Most numerical columns are clean and realistic, with some skew in income-related features.
* Credit_Score is imbalanced, with majority in “Standard” category.
* Customers can have multiple loans, captured through Loan_Count.
* Data reflects financial behavior patterns like spending, delays, and credit usage.


## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1 (Delay from Due Date vs Credit Score) (Boxplot)

````python
# Chart - 1 visualization code
sns.boxplot(x="Credit_Score", y="Delay_from_due_date", data=df)

plt.xlabel("Credit Score")
plt.ylabel("Delay from Due Date")
plt.title("Delay from Due Date vs Credit Score")

plt.show()
````

##### 1. Why did you pick the specific chart?

Since delay is a continuous variable and credit score is categorical, a boxplot is suitable to visualize their relationship.

##### 2. What is/are the insight(s) found from the chart?

* Customers with higher delay from due date tend to have poorer credit scores.
* Good credit score customers have the lowest median delay, indicating timely payments.
* Standard customers show moderate delay, indicating occasional late payments.
* Poor credit score customers have the highest delay and wider spread, showing inconsistent and risky payment behavior.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, the insights can create a positive business impact. Since higher payment delays are strongly linked to poorer credit scores, financial institutions can use this to identify high-risk customers early. This allows them to implement stricter credit policies, adjust interest rates, or send timely payment reminders, which can reduce defaults and improve overall portfolio quality.
